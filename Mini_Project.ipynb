```python
858484# Instructors Management System
instructors = [
    {'name': 'Fatima Malik', 'subject': 'WordPress Designer & Developer|Physicist',
     'address': 'Okara District, Punjab, Pakistan', 'contact': '7223fatimamalik@gmail.com',
     'linkedin': 'https://www.linkedin.com/in/fatima-malik99/', 'github': 'https://github.com/fatima-malik99',
     'classes': 'Monday, Wednesday, Friday'},
    {'name': 'Sana Raza', 'subject': 'Biomedical Engineer | Technical Content Writer',
     'address': 'Karāchi, Sindh, Pakistan', 'contact': 'sana.raxa321@gmail.com',
     'linkedin': 'https://www.linkedin.com/in/sana-raza-engineer/', 'github': 'https://github.com/Sanarazaaa',
     'classes': 'Monday, Thursday, Saturday'},
    {'name': 'Ubaid Ur Rehman', 'subject': 'Software Engineer | Section Leader',
     'address': 'Islāmābād, Pakistan', 'contact': 'ubaid_ur_Rehman@gmail.com',
     'linkedin': 'https://www.linkedin.com/in/iubaidrmn/', 'github': 'https://github.com/',
     'classes': 'Tuesday, Thursday'},
    {'name': 'Laiba Asif', 'subject': 'Python | AI Enthusiast | Data analyst',
     'address': 'Jhang District, Punjab, Pakistan  ', 'contact': 'laibasif99@gmail.com',
     'linkedin': 'https://www.linkedin.com/in/laiba-asif-8544a0277/', 'github': 'https://github.com/',
     'classes': 'Tuesday, Thursday'}
]

def display_instructors():
    print('\nOur Beloved Python For Beginners  Instructors:')
    s_n = 1
    for instructor in instructors:
        print(f"{s_n}. {instructor['name']}")
        s_n += 1

def view_instructors_info():
    try:
        instructors_number = int(input('Enter the instructor number (1-4) about want to know : ')) - 1
        if 0 <= instructors_number < len(instructors):
            instructor = instructors[instructors_number]
            print(f"\nName: {instructor['name']}")
            print(f"Subject: {instructor['subject']}")
            print(f"Address: {instructor['address']}")
            print(f"Contact: {instructor['contact']}")
            print(f"LinkedIn: {instructor['linkedin']}")
            print(f"Github: {instructor['github']}")
            print(f"Classes: {instructor['classes']}")
        else:
            print("Invalid instructor number!")
    except ValueError:
        print("Please enter a valid number.")

def add_instructors():
    name = input("Enter the instructor's name: ")
    subject = input("Enter the subject specialty of the instructor: ")
    address = input("Enter the instructor's address: ")
    contact = input("Enter the instructor's contact email: ")
    linkedin = input("Enter the instructor's LinkedIn link: ")
    github = input("Enter the instructor's GitHub link: ")
    classes = input("Enter the instructor's classes (e.g., 'Monday, Wednesday'): ")

    # Add the new instructor
    instructors.append({
        'name': name,
        'subject': subject,
        'address': address,
        'contact': contact,
        'linkedin': linkedin,
        'github': github,
        'classes': classes,
    })
    print(f"\nNew instructor {name} added successfully!")

def update_instructors():
    try:
        instructors_number = int(input("Enter the instructor number (1-3) to update: ")) - 1
        if 0 <= instructors_number < len(instructors):
            instructor = instructors[instructors_number]
            print(f"\nUpdating details for {instructor['name']}:")

            # Update details
            instructor['name'] = input(f"Enter new name (current: {instructor['name']}): ") or instructor['name']
            instructor['subject'] = input(f"Enter new subject (current: {instructor['subject']}): ") or instructor['subject']
            instructor['address'] = input(f"Enter new address (current: {instructor['address']}): ") or instructor['address']
            instructor['contact'] = input(f"Enter new contact email (current: {instructor['contact']}): ") or instructor['contact']
            instructor['linkedin'] = input(f"Enter new LinkedIn link (current: {instructor['linkedin']}): ") or instructor['linkedin']
            instructor['github'] = input(f"Enter new GitHub link (current: {instructor['github']}): ") or instructor['github']
            instructor['classes'] = input(f"Enter new classes (current: {instructor['classes']}): ") or instructor['classes']

            print(f"\nInstructor details updated successfully!")
        else:
            print("Invalid instructor number!")
    except ValueError:
        print("Please enter a valid number.")

# Main program
def main():
    while True:
        print("\nInstructors Management System (IMS)")
        print("1. View all instructors")
        print("2. View instructor details")
        print("3. Add a new instructor")
        print("4. Update instructor details")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            display_instructors()
        elif choice == '2':
            view_instructors_info()
        elif choice == '3':
            add_instructors()
        elif choice == '4':
            update_instructors()
        elif choice == '5':
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the main program
if __name__ == "__main__":
    main()
```
